```python

import pandas as pd


base_name = "ecoli"

label_mapping = {'cp': 0, 'im':1, 'pp':2, 'imU':3, 'om':4, 'omL':5, 'imL':6, 'imS':7}

for i in range(1,6):
    print(f"Transform-{i}")
    file_name = f"{base_name}-5-{i}"
    
    data_train = pd.read_csv( f"{base_name}/{base_name}-5-fold/{file_name}tra.dat"  ,
        header=None, 
        delimiter=',', 
        comment='@',
        skipinitialspace=True  # Remove espaços após vírgulas
    ) 
    data_test = pd.read_csv( 
        f"{base_name}/{base_name}-5-fold/{file_name}tst.dat"  , 
        header=None, 
        delimiter=',', 
        comment='@', 
        skipinitialspace=True  # Remove espaços após vírgulas
    ) 

    data_train = data_train.replace(label_mapping, regex=True)
    
    data_test = data_test.replace(label_mapping, regex=True)

    
    # Salvar
    data_train.to_csv(f"{base_name}/{base_name}-transformed/{file_name}tra.dat", index=False, header=False)

 
    # Salvar
    data_test.to_csv(f"{base_name}/{base_name}-transformed/{file_name}tst.dat", index=False, header=False)
```
